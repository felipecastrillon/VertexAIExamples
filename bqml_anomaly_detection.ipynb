# Analyzing Anomalies with Time Series Data Using BQML

### Get Top 10 Station Names Based on # of Trips

```python
%%bigquery 
SELECT start_station_name, count(*) as trips
FROM `bigquery-public-data.new_york_citibike.citibike_trips`
WHERE date(starttime) between "2017-06-26" AND "2017-07-26"
GROUP BY start_station_name
ORDER BY trips DESC
```

### Get Training Data and Testing Data for The Top Ten Stations

Get 3 months of training data and 1 month of testing data. Only select data within the top 10 stations based on total trips.

```python
%%bigquery
CREATE OR REPLACE TABLE `felipe-sandbox-354619.time_series.citibike_train_data` AS
SELECT start_station_name, date(starttime) as start_date, count(*) as trips
FROM `bigquery-public-data.new_york_citibike.citibike_trips` 
WHERE date(starttime) between "2018-01-01" AND "2018-03-31"
AND start_station_name IN ( 
    SELECT start_station_name
    FROM `bigquery-public-data.new_york_citibike.citibike_trips`
    WHERE date(starttime) between "2018-01-01" AND "2018-03-31"
    GROUP BY start_station_name
    ORDER BY count(*) DESC
    LIMIT 10
) 
GROUP BY start_station_name, start_date
```

```python
%%bigquery
SELECT count(*) FROM `felipe-sandbox-354619.time_series.citibike_train_data`
```

```python
%%bigquery
CREATE OR REPLACE TABLE `felipe-sandbox-354619.time_series.citibike_test_data` AS
SELECT start_station_name, date(starttime) as start_date, count(*) as trips
FROM `bigquery-public-data.new_york_citibike.citibike_trips` 
WHERE date(starttime) between "2018-04-01" AND "2018-04-30"
AND start_station_name IN ( 
    SELECT start_station_name
    FROM `bigquery-public-data.new_york_citibike.citibike_trips`
    WHERE date(starttime) between "2018-04-01" AND "2018-04-30"
    GROUP BY start_station_name
    ORDER BY count(*) DESC
    LIMIT 10
) 
GROUP BY start_station_name, start_date
```

```python
%%bigquery
SELECT count(*) FROM `felipe-sandbox-354619.time_series.citibike_test_data`
```

### Train ARIMA PLUS Model for Each Station

```python
%%bigquery
CREATE OR REPLACE MODEL
  `felipe-sandbox-354619.time_series.citibike_arima`
OPTIONS(
  model_type='ARIMA_PLUS',
  holiday_region = 'US',
  time_series_timestamp_col='start_date',
  time_series_data_col='trips',
  time_series_id_col='start_station_name'
) AS
SELECT start_date,start_station_name,trips FROM `felipe-sandbox-354619.time_series.citibike_train_data`
```

```python
%%bigquery
SELECT * FROM ML.ARIMA_EVALUATE(MODEL `felipe-sandbox-354619.time_series.citibike_arima`, STRUCT(FALSE AS show_all_candidate_models))
```

All the stations have weekly seasonality and "Broadway & E 14 st" has some spikes and dips which could potentially be anomalies. We will investigate later. 

### Forecast 30 Data Point Into the Future and Join Results With Ground Truth Data

Join forecast results with test data to compare forecast vs ground truth. Also, concatenate with training data to view full range of dates covering the training and testing data

```python
%%bigquery 
CREATE OR REPLACE TABLE `felipe-sandbox-354619.time_series.citibike_forecast_results` AS
SELECT a.start_station_name, 
        date(a.forecast_timestamp) as start_date, 
        a.forecast_value as forecasted_trips, 
        a.prediction_interval_lower_bound as lower_bound, 
        a.prediction_interval_upper_bound as upper_bound,
        b.trips
FROM #forecast data
    ML.FORECAST(MODEL `felipe-sandbox-354619.time_series.citibike_arima`,
    STRUCT(30 AS horizon, 0.95 AS confidence_level)) AS a
JOIN (
    #test data
    SELECT *, null as forecasted_trips, null as lower_bound, null as upper_bound FROM `felipe-sandbox-354619.time_series.citibike_test_data` 
) AS b
ON CONCAT(a.start_station_name, date(a.forecast_timestamp)) = CONCAT(b.start_station_name, b.start_date)

UNION ALL

# train data
SELECT
    *, 
    null as forecasted_trips, 
    null as lower_bound, 
    null as upper_bound
FROM
`felipe-sandbox-354619.time_series.citibike_train_data` 
```

### Quick Discussion on Prediction Bands and Anomalies

The first station "Pershing Square North" seems to follow a very predictable weekly pattern and thus the predictions it makes have a tighter prediction band (in green). And not surprisingly, all real values fall inside the bands except for maybe one value at the beginning of the time series. On the other hand, the second station "Broadway & E 14 St" does not have a very predictable pattern and thus the model has a broad prediction band without too much seasonality. What is really surprising is that even with the conservatism of the forecast the actual trips still spike outside of the confidence bands. This is a clear anomaly point.

### Using the Built-in Anomaly Detection Call
One way to evaluate anomalies in BQML is to detect the values that fall oustide the prediction bounds. However, an easier approach is to use the built-in anomaly detection calls. Let's try it.

```python
%%bigquery 
CREATE OR REPLACE TABLE `time_series.citibike_anomalies` AS
SELECT
  *
FROM
  ML.DETECT_ANOMALIES(MODEL `time_series.citibike_arima`,
                      STRUCT(0.95 AS anomaly_prob_threshold), 
                     (select * from felipe-sandbox-354619.time_series.citibike_test_data));
```

```python
%%bigquery df
SELECT * FROM `time_series.citibike_anomalies`
```

```python
df = df.sort_values(by="start_date")
df = df.set_index("start_date")
df.head()
```

```python
import numpy as np

# Create the time series plot
df_one = df[df["start_station_name"]=="Pershing Square North"]

# Define a color mapping for categories
df_one["is_anomaly"] = df_one["is_anomaly"].fillna(False)
df_one['colors'] = np.where(df_one["is_anomaly"] == True, "red", "blue")

plt.figure(figsize=(10, 4))
plt.plot(df_one["trips"], marker="o", label="Trips")
plt.plot(df_one[df_one["is_anomaly"]==True]["trips"], marker="o", linestyle="", label="Anomalies")

plt.xlabel("Date")  # Adjust label if needed
plt.ylabel("Number of Trips")
plt.title("Time Series of Trips with Anomalies for Pershing Square North")
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
