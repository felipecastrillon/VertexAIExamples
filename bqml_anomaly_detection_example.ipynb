# Analyzing Anomalies with Time Series Data Using BQML

### Get Top 10 Stores based on Sales

```python
%%bigquery 
SELECT store_name, SUM(sale_dollars) AS total_sales
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN DATE_SUB(CURRENT_DATE(), INTERVAL 1 MONTH) AND CURRENT_DATE()
GROUP BY store_name
ORDER BY total_sales DESC
LIMIT 10;
```

### Get Training Data and Testing Data for The Ten Stores

```python
%%bigquery
CREATE TABLE `solutions-2023-mar-107.`
SELECT *
FROM `bigquery-public-data.iowa_liquor_sales.sales` a
WHERE date BETWEEN DATE_SUB(CURRENT_DATE(), INTERVAL 1 MONTH) AND CURRENT_DATE()
AND store_name IN (
    SELECT store_name
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date BETWEEN DATE_SUB(CURRENT_DATE(), INTERVAL 1 MONTH) AND CURRENT_DATE()
GROUP BY store_name
ORDER BY SUM(sale_dollars) DESC
LIMIT 10) 
```
