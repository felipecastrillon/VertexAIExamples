```python
%%bigquery

CREATE SCHEMA bqml_timeseries
  OPTIONS (
    default_table_expiration_days = 5,
    location = 'us'
  );
```

```python
%%bigquery

CREATE TABLE `bqml_timeseries.seattle_air_quality_daily`
AS
WITH
  pm25_daily AS (
    SELECT
      avg(arithmetic_mean) AS pm25, date_local AS date
    FROM
      `bigquery-public-data.epa_historical_air_quality.pm25_nonfrm_daily_summary`
    WHERE
      city_name = 'Seattle'
      AND parameter_name = 'Acceptable PM2.5 AQI & Speciation Mass'
    GROUP BY date_local
  ),
  wind_speed_daily AS (
    SELECT
      avg(arithmetic_mean) AS wind_speed, date_local AS date
    FROM
      `bigquery-public-data.epa_historical_air_quality.wind_daily_summary`
    WHERE
      city_name = 'Seattle' AND parameter_name = 'Wind Speed - Resultant'
    GROUP BY date_local
  ),
  temperature_daily AS (
    SELECT
      avg(first_max_value) AS temperature, date_local AS date
    FROM
      `bigquery-public-data.epa_historical_air_quality.temperature_daily_summary`
    WHERE
      city_name = 'Seattle' AND parameter_name = 'Outdoor Temperature'
    GROUP BY date_local
  )
SELECT
  pm25_daily.date AS date, pm25, wind_speed, temperature
FROM pm25_daily
JOIN wind_speed_daily USING (date)
JOIN temperature_daily USING (date)
```

```python
%%bigquery

CREATE TABLE `bqml_timeseries.seattle_air_quality_daily_train` AS

SELECT * FROM `bqml_timeseries.seattle_air_quality_daily`
WHERE date <= date("2020-12-31")
```

```python
%%bigquery

CREATE OR REPLACE
  MODEL
    `bqml_timeseries.seattle_pm25_xreg_model`
  OPTIONS (
    MODEL_TYPE = 'ARIMA_PLUS_XREG',
    time_series_timestamp_col = 'date',
    time_series_data_col = 'pm25')
AS
SELECT
    *
FROM
  `bqml_timeseries.seattle_air_quality_daily_train`
```

```python
%%bigquery

SELECT
  *
FROM
  ML.FORECAST(
    MODEL
      `bqml_timeseries.seattle_pm25_xreg_model`,
    STRUCT(30 AS horizon),
    (
      SELECT
        date,
        temperature,
        wind_speed
      FROM
        `bqml_timeseries.seattle_air_quality_daily`
      WHERE
        date > DATE('2020-12-31')
    ))
```

```python
%%bigquery

SELECT * FROM `bqml_timeseries.seattle_air_quality_daily_train`
```

```python
%%bigquery df

CREATE TABLE `bqml_timeseries.evaluation` AS

SELECT
  *
FROM
  ML.EVALUATE(
    MODEL `bqml_timeseries.seattle_pm25_xreg_model`,
    (
      SELECT
        date as date,
        pm25,
        temperature,
        wind_speed
      FROM
        `bqml_timeseries.seattle_air_quality_daily`
      WHERE
        date > DATE('2020-12-31')
    ),
    STRUCT(
      FALSE AS perform_aggregation,
      30 AS horizon)) as eval_table

UNION ALL

SELECT
    timestamp(date) as date, 
    pm25 as pm25, 
    NULL as forecasted_pm25,
    NULL as lower_bound,
    NULL as upper_bound,
    NULL as absolute_error,
    NULL as absolute_percentage_error,
    temperature as temperature,
    wind_speed as wind_speed
FROM
`bqml_timeseries.seattle_air_quality_daily_train` 
WHERE date > date('2020-10-01')
;
```
